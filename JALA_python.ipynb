```python
def find_duplicates(arr):
    duplicates=[]
    for i in arr:
        if arr.count(i) > 1 and i not in duplicates:
            duplicates.append(i)
    return duplicates
print(find_duplicates([1, 2, 3, 2, 4, 1]))
```

```python
def remove_duplicates(arr):
    return list(set(arr))
print(remove_duplicates([1,2,2,3,4,4]))
```

```python
def second_largest(arr):
    arr=list(set(arr))
    arr.sort()
    return arr[-2]
print(second_largest([10,20,30,40,50]))
```

```python
def second_largest_alt(arr):
    first = second = float('-inf')
    for num in arr:
        if num > first:
            second=first
            first=num
        elif num > second and num!=first:
            second=num
    return second
print(second_largest_alt([10, 20, 30, 40, 50]))
```

```python
def count_even_odd(arr):
    even=odd=0
    for num in arr:
        if num%2==0:
            even += 1
        else:
            odd += 1
    return even, odd
print(count_even_odd([1,2,3,4,5,6]))
```

```python
def difference(arr):
    return max(arr) - min(arr)
print(difference([10, 5, 20, 8]))
```

```python
def contains_both(arr):
    return 12 in arr and 23 in arr
print(contains_both([12, 5, 23, 9]))
```

```python
def remove_duplicates_new(arr):
    new_arr=[]
    for num in arr:
        if num not in new_arr:
            new_arr.append(num)
    return new_arr
print(remove_duplicates_new([1,2,2,3,4,4]))
```

7. INHERITANCE

```python
class A:
    def show_A():
        print("This is class A")

    def display():
        print("This is again  class A")


class B(A):
    def show_B():
        print("This is class B")

    def display():
        print("This is again  class B")
class C(B):
    def show_C():
        print("This is class C")

    def display():
        print("This is again  class C")


obj1 = A
obj2 = B
obj3 = C

print(obj3.show_A()) # inherited method
print(obj3.display()) #method over ride 
print(obj3.show_B())
```

8. ACCESS MODIFIERS

```python
class A:
    __x=10
    __y=20
    def __private_method(self):
        print("Private method inside class A")
    def main(self):
        print("Private field x:",self.__x)
        print("Private field y:",self.__y)
        self.__private_method()
class B(A):
    def access(self):
        try:
            print(self.__x)
        except:
            print("Cannot access private field from subclass")
        try:
            self.__private_method()
        except:
            print("Cannot access private method from subclass")
a=A()
a.main()
b=B()
b.access()
```

```python
class A:
    _x=100
    _y=200
    def _protected_method(self):
        print("Protected method inside class A")

class B:
    def access(self):
        obj=A()
        print("Accessing protected field:",obj._x)
        obj._protected_method()

b=B()
b.access()

```

```python
class C(A):
    def access_child(self):
        print("Accessing protected field from child:",self._x)
        self._protected_method()

c=C()
c.access_child()
```

```python
class A:
    x = 40
    y = 70
    def public(self):
        print("public method inside class A")
class B:
    def access(self):
        obj=A()
        print("accessing public field",obj.x)
        obj.public()
b=B()
b.access()
```

9. ABSTRACT CLASS

```python
from abc import ABC,abstractmethod
class A(ABC):
    def method(self):
        print("this is not an abstract class")
    def method2(self):
        pass
```

```python
from abc import ABC,abstractmethod
class A(ABC):
    def normal_method(self):
        print("Non-abstract method in A")
    @abstractmethod
    def abstract_method(self):
        pass
class B(A):
    def abstract_method(self):
        print("Abstract method implemented in B")
    def access(self):
        ref=self
        ref.normal_method()
obj=B()
obj.access()
```

```python
from abc import ABC,abstractmethod
class A(ABC):
    @abstractmethod
    def abstract_method(self):
        pass
class B(A):
    def abstract_method(self):
        print("Abstract method implemented in B")
    def create_instance(self):
        obj=B()
        obj.abstract_method()
b=B()
b.create_instance()
```

```python
from abc import ABC,abstractmethod
class A(ABC):
    def normal_method(self):
        print("Non-abstract method in A")
    @abstractmethod
    def abstract_method(self):
        pass
class B(A):
    def abstract_method(self):
        print("Abstract method implemented in B")
    def create_instance(self):
        obj=B()
        obj.normal_method()
b=B()
b.create_instance()
```

```python
class A :
    def __init__(self,a = None):
        if a is None:
            print("super default constructor")
        else:
            print("super argument constructor",a)
class B(A):
    def __init__(self,a=None) :
        super().__init__(a)
        print("child constructor")
obj1=B()
obj2=B(100)
```

```python
class A:
    def __init__(self,a=None):
        print("constructor")
    def __private(self):
        print("Private constructor")
    def public(self):
        print("public constructor")
    def _protected(self):
        print("protected constructor")
obj=A()
obj._protected()
obj.public()
```
