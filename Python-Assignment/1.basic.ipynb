Define the local and Global variables with the same name and print both variables and
understand the scope of the variables

```python
#user score
score = 0

def comp_score():
    score = 5
    print(f"Computer score is: {score}") # should display 5


def user_score():
    global score
    score = 2
    print(f"User score is: {score}") #should display 2

comp_score()
user_score()
```
