5. Write a program to read a file a just to a particular index using seek()

```python
with open("files/file2.txt", "r") as fp:

    print("first 10 char: \n")
    print(fp.read(10))

    #pointer is at 10th position
    print("pointer position" , fp.tell())

    print("Shifting to pointer")
    fp.seek(20)
    print(fp.read(10))

    print("pointer position" , fp.tell())
```
