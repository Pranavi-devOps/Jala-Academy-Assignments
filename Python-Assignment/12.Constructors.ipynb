2. Call the constructors(both default and argument constructors) of super class from a child
class

```python
class A:

   
    def __init__(self, name=None):
        if name is None:
            print("default constructor of a")
            self.name = "kartikey"
        else:
            print("other constructor of a")
            self.name = name

    def print_A(self):
        print("Name from A:", self.name)


class B(A):

    def __init__(self, name=None):
        print("constructor of b")
        super().__init__(name)   #constructor of a 
    def print_B(self):
        print("Name from B:", self.name)


obj1 = B()          
obj1.print_A()

obj2 = B("Mohasin")      
obj2.print_B()
```

```python
class C:

  
    def __init__(self, name, roll, branch):
        self.name = name #public

        self._roll = roll #protected

        self.__branch = branch #protected 

    def displayName(self):
        print("Name:", self.name)

    def _displayRoll(self):
        print("roll no :", self._roll)


    def __displayBranch(self):
        print("branch:", self.__branch)

    def accessBranch(self):
        self.__displayBranch()


class D(C):

    def __init__(self, name, roll, branch):
        super().__init__(name, roll, branch)

    def accessRoll(self):
        self._displayRoll() 

obj3 = D("kartikey", 90, "CSD")

obj3.displayName()    
obj3.accessRoll()      
obj3.accessBranch()
```
