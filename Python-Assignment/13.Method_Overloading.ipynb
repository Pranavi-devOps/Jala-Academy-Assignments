1. Write two methods with the same name but different number of parameters of same type
and call the methods

```python
#you cannot method overload in python like in java

def display(name ,age = None):
    if age == None:
        print("Name",name)
    else:
        print("Name",name)
        print("age",age)



display(name="srikanth",age=19)   
display("kartikey")
```

2. Write two methods with the same name but different number of parameters of different
data type and call the methods 

```python
def display(name ,age = None):
    if age == None:
        print("Name",name)
    else:
        print("Name",name)
        print("age",float(age))



display(name="srikanth",age=19)   
display("kartikey")
```
