4. Write a program with multiple catch blocks

```python
try:
    a  = int(input())
    b = int(input())
    print(a/b)
except ValueError:
    print("Enter an int")
except ZeroDivisionError:
    print("cannot divide by zero")
except exception as e:
    print("Unexpected error")
```

7. Write a program with finally block

```python
try:
    def divide():
        a  = int(input())
        b = int(input())
        print(a/b)
    
    divide()
except ValueError:
    print("Enter an int")
except ZeroDivisionError:
    print("cannot divide by zero")
except exception as e:
    print("Unexpected error")
finally:
    divide()
```
