4. Program for relational operators (<,<==, >, >==)

```python
a = int(input("Enter first number: \n"))
b = int(input("Enter second number: \n"))

print("a>b",a>b) #will return a bool value
print("a>=b",a>=b)
print("a<b",a<b)
print("a<=b",a<=b)
```

5. Print the smaller and larger number

```python
a = int(input("Enter first number: \n"))
b = int(input("Enter second number: \n"))

if a > b :
    print("a is larger")
else:
    print("b is larger")

if a < b:
    print("a is smaller")
else:
    print("b is smaller")
```
