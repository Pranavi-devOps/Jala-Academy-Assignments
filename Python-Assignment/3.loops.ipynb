8. Write a program to find Armstrong number or not

```python
# To check if a number is armstrong or not it should be digit * digit * digit = the number
# easiest way to do this is by using strings and converting them into numbers

num = int(input("Enter a number: \n"))

temp = num 
armstrong_num = 0

while temp != 0:
    last_digit = temp % 10
    armstrong_num = armstrong_num + (last_digit*last_digit*last_digit)
    temp = temp // 10

if armstrong_num == num:
    print("Armstrong number")
else:
    print("Not an armstrong number")
```

9. Write a program to find the prime or not.


```python
is_prime = True

n = int(input("Enter a number: \n"))

for num in range(2,(n//2)): # we use 2 to avoid gettin divided by 1 and we use n/2 as nothing over that number can be divisible by n 
    if n % num == 0:
        is_prime = False
        break

if is_prime == True:
    print(f"{n} is Prime")
else:
    print(f"{n} is not prime")
```

10. Write a program to palindrome or not.

```python
n = input("Enter a number: \n")
print("palindrome" if n == n[::-1] else "not palindrome")
# above is the easiest way to do it 
```

```python
num = int(input("Enter a number: \n"))

temp = num 
rev = 0

while temp != 0:
    last_digit = temp % 10
    rev = (rev * 10 ) + last_digit
    temp = temp // 10

if rev == num:
    print("palindrome number")
else:
    print("Not an palindrome number")
```

11. Program to check whether a number is EVEN or ODD using switch

```python
#switch cases are called match in python 

n = int(input("Enter a number:\n"))

match n:
    case n if n % 2 ==0:
        print("Even")
    case n if n % 2 !=0:
        print("odd")

```
