3. Write a program to find the index of an array element

```python
#the simplest way to do it without any inbuild functions is linear search

def linear_search(arr , key):

    for i in range (0,len(arr)):

        if arr[i] == key:
            return i
        
    return -1



arr = list(map(int,input().split()))
key = int(input("Enter the value you are searching for: "))

result = linear_search(arr,key)

if result == -1:
    print("Element not found")
else:
    print(f"Element found at index :{result}")
```

5. Write a function to remove a specific element from an array

```python
#using the same linear search we should just remove the specific value instead of returning the index 

def linear_search(arr , key):

    for i in range (0,len(arr)):
        
        if arr[i] == key:
            arr.pop(i) #uses index to pop the item
            return 1
        
    return -1



arr = list(map(int,input().split()))
key = int(input("Enter the value you want to remove: \n "))

result = linear_search(arr,key)


if result == -1:
    print("Element was not in the array")
else:
    print(f"The element {key} was removed from the array")
    print(arr)
```

8. Write a function to find the minimum and maximum value of an array


```python
#again , using the min and max functin is the easiest way to do this problem but i will do it without them 

def largest(arr):
    largest = arr[0]
    for num in arr:
        if num > largest:
            largest = num
        
    return largest


def smallest(arr):
    smallest = arr[0]
    for num in arr:
        if num < smallest:
            smallest = num
        
    return smallest

arr = list(map(int,input().split()))
larg = largest(arr)
small = smallest(arr)

print(f"The smallest number is: {small}")
print(f"The largest number is: {larg}")
```

11. Write a program to find the common values between two arrays

```python
# we can convert the lists into sets and then find the common by intersection method 

arr = list(map(int,input().split()))
arr2 = list(map(int,input().split()))

set1 = set(arr)
set2 = set(arr2)
print(arr)
print(arr2)

print(f"The common values are: {set1.intersection(set2)}")
```

13. Write a method to find the second largest number in an array

```python
#13 and 14th are the same question where we need to first find largest and then replace it by 2nd largest

arr = list(map(int,input().split()))

largest = arr[0]
second_largest = arr[0]

for num in arr:
    if num > largest:
        second_largest = largest
        largest = num 

    
print(f"Largest number : {largest}")
print(f"Second largest: {second_largest}")
```

16. Write a function to get the difference of largest and smallest value


```python
def difference(arr):
    largest = max(arr)
    smallest = min(arr)

    return largest - smallest


arr = list(map(int,input().split()))
resutl = difference(arr)

print(f"The difference between the largest and the smallest is : {resutl}")
```

17. Write a method to verify if the array contains two specified elements(12,23)

```python
def verify(arr):

    twelve = False
    twentry_three = False

    for num in arr:
        if num == 12:
            twelve = True
        if num == 23:
            twentry_three = True

    if twentry_three  == True and twelve == True:
        return True
    else:
        return False
    

arr = list(map(int,input().split()))

result = verify(arr)

if result == True:
    print("The list has 12 and 23")
else:
    print("The list does not has 12 and 23")
```
