1. Define a static variable and access that through a class 

```python
class Student:
    age = 19 #in python we dont need to use the static keyword to define a static variable 


print(Student.age)
```

2. Define a static variable and access that through a instance

```python
#instance means we need to make an object 
class Student:
    age = 19

obj = Student()

print(obj.age)
```

3. Define a static variable and change within the instance

```python
class Student:
    age = 19

obj = Student()

obj.age = 22
print(obj.age)
```

4. Define a static variable and change within the class

```python
class Student:
    age = 19

obj = Student()

obj.age = 22
print(obj.age)
print(Student.age)
```
