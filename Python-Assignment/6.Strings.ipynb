1. Different ways creating a string


```python
#there are 3 ways to create a string using 

print("hello world")

print('hello world')

print("""
hello 
world
""")
```

2. Concatenating two strings using + operator

```python
print("hello " + "World")
```

3. Finding the length of the string

```python
str = "So there was this one time in this imaginary world there was a little guy name kittu"

print(len(str))
```

4. Extract a string using Substring

```python
str = "Good bye world"

print(str[0:4])  # 0 is included and 4 is excluded
```

5. Searching in strings using index()

```python
str = "So there was this one time in this imaginary world there was a little guy name kittu"

print(str.index("was")) # displays the index of W 
```

7. Comparing strings

```python
str = "Good bye world"
str2 = "Good bye world"

print(str == str2)
print(str != str2)
```

8. startsWith(), endsWith() and compareTo()


```python
str = "Good bye world"

print(str.startswith("goo")) # Return a bool value and is case sensitive 
print(str.endswith("world"))
#there is no compare to command in python 
```

9. Trimming strings with strip()

```python
# this removed teh spaces from a string 

str = "       hellow     guyyys   "
print(str.strip())
```

10. Replacing characters in strings with replace()

```python
str = "Good bye world"

print(str.replace("Good bye" , "hello"))
```

11. Splitting strings with split()

```python
str = "So there was this one time in this imaginary world there was a little guy name kittu"

print(str.split()) # splits the string into a list of words 
```

12. Converting integer objects to Strings

```python
n = int(input("Enter an integer"))

str3 = str(n)
print(str3) # the code is correct but is probably taking str the variable and not the data type 
```

13. Converting to uppercase and lowercase
