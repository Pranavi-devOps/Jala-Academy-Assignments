Create three methods in each class, 2 methods are specific to each class and third
method (override method) should be in all three Classes A, B and C


```python
class A:
    def show_A():
        print("This is class A")

    def display():
        print("This is again  class A")


class B(A):
    def show_B():
        print("This is class B")

    def display():
        print("This is again  class B")
class C(B):
    def show_C():
        print("This is class C")

    def display():
        print("This is again  class C")


obj1 = A
obj2 = B
obj3 = C

print(obj3.show_A()) # inherited method
print(obj3.display()) #method over ride 
print(obj3.show_B())
```

```python
class A:
    def show_A():
        print("This is class A")

    def display():
        print("This is again  class A")


class B(A):
    def show_B():
        print("This is class B")

    def display():
        print("This is again  class B")
class C(B):
    def __init__(self):
        super().__init__()
    def show_C(self):
        print("This is class C")

    def display(self):
        print("This is again  class C")


obj1 = A
obj2 = B
obj3 = C

print(obj3.display()) #over ridden 
print(obj3.display())
```
