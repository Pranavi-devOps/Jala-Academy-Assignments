1. Create an abstract class with abstract and non-abstract methods.

2. Create a sub class for an abstract class. Create an object in the child class for the
abstract class and access the non-abstract methods


```python
from abc import ABC, abstractmethod
#abc is used to import abstract methods in python
```

```python
class animal(ABC): #this is an abstract class

    @abstractmethod
    def sound(self):
        pass

    def sleep(self):
        print("animal is sleeping")


class dog(animal):

    def sound(self):
        print("BHAW BHAW ")

#obj1 = animal()
#obj1.sleep()
#the code above will show an error 

obj2 = dog()
obj2.sound()
obj2.sleep()
```
